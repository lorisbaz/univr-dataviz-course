# Line plots

**Purpose**: Showing trends in data -- usually time series data with many time points.

**matplotlib function**: [`plot(x, y)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)

* `x`: The x-coordinates of the lines or markers.

* `y`: The y-coordinates of the lines or markers.

```python
x = np.linspace(0, 10, 100)
```

**Commonly used parameters**:

* `color`: Set the color of the line.

* `linestyle`: Set the line style, e.g., solid, dashed, or none.

* `linewidth`: Set the line thickness.

* `marker`: Set the marker style, e.g., circles, triangles, or none.

* `markersize`: Set the marker size.

* `label`: Set the label for the line that will show up in the legend.

```python
plt.plot(x, np.sin(x - 0))        
plt.plot(x, np.sin(x - 1))       
plt.plot(x, np.sin(x - 2))        
```

```python
x = np.linspace(0, 20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)

plt.figure()

plt.plot(x,
         y1,
         color='black',
         linestyle='-',
         linewidth=2,
         marker='s',
         markersize=6,
         label='y1')

plt.plot(x,
         y2,
         color='gray',
         linestyle='--',
         linewidth=2,
         marker='^',
         markersize=6,
         label='y2')

plt.legend()
;
```

```python
plt.plot(x, np.sin(x - 0), color='blue', linestyle='-')  # 'solid' does the same
plt.plot(x, np.sin(x - 1), color='g', linestyle='--') # dashed
plt.plot(x, np.sin(x - 2), color='0.75', linestyle='-.') # dashdot
plt.plot(x, np.sin(x - 3), color='#FFDD44', linestyle=':');  # dotted
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3), linewidth=6,
         linestyle=(0, (5, 2, 1, 2)), dash_capstyle='round')
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
, linestyle='-')  # solid
```

```python
t = np.arange(0., 10., 0.5)
```

```python
#We can put some additional items
plt.plot(t, np.sin(t), 'ro', t, 2*np.sin(t), 'bs', t, 3*np.sin(t), 'g^')
plt.annotate('local max', xy=(2, 3), xytext=(4, 3),
             arrowprops=dict(facecolor='black', shrink=0.08),
             )


#points, different markers
plt.savefig('filename'); 
plt.show()
```

# Vertical bar charts

**Purpose**: Comparing categories **OR** showing temporal trends in data with few (< 4) time points.

**matplotlib function**: [`bar(left, height)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar)

* `left`: The x coordinate(s) of the left sides of the bars.

* `height`: The height(s) of the bars.

```python
import numpy as np
import matplotlib.pyplot as plt

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#we can play with bars position (and use multiple bars)
years = np.arange(2012, 2015)
category1_values = [2, 5, 9]
category2_values = [7, 6, 3]

plt.figure()

plt.bar(years - 0.2,
        category1_values,
        color='blue',
        edgecolor='none',
        width=0.4,
        align='center',
        label='y1')

plt.bar(years + 0.2,
        category2_values,
        color='orange',
        edgecolor='none',
        width=0.4,
        align='center',
        label='y2')

plt.xticks(years, [str(year) for year in years])

plt.legend()
```

# Horizontal bar charts

**Purpose**: Comparing categories.

**matplotlib function**: [`barh(bottom, width)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh)

* `bottom`: The y coordinate(s) of the bars.

* `width`: The width(s) of the bars.

```python
categories = ['A', 'B', 'C', 'D', 'E','F','G','H','I','L','M']
values = [7, 12, 4, 2, 9,4,5,3,9,11,5]
my_cmap = plt.get_cmap("tab20")
plt.figure()

plt.barh(np.arange(len(categories)), values, color=my_cmap.colors)
plt.yticks(np.arange(len(categories)),
           ['Category {}'.format(x) for x in categories])
```

# stacked bar charts

**Purpose**: Comparing values with two categories

```python
# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

# Pie charts

**Purpose**: Displaying a simple proportion.

**matplotlib function**: [`pie(sizes)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie)

* `sizes`: The size of the wedges as either a fraction or number.

**Commonly used parameters**:

* `colors`: Set the colors of the wedges.

* `labels`: Set the labels of the wedges.

* `startangle`: Set the angle that the wedges start at.

* `autopct`: Set the percentage display format of the wedges.
    

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov= pd.read_csv('data/types_movies.csv')

plt.rcdefaults()

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'])

plt.axis('equal')

plt.show()
```

```python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered','chocolate']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f')

plt.axis('equal')

plt.show()
```

```python
explode = (0, 0.1, 0, 0, 0.3)

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```

# Heatmap
**Purpose**:  visualize one ordinal value as a functions of two keys
** note:** is the same thing we do in 2D image visualization (we'll see), but we typically limited number of key values, not usually ordered, and annotades

```python
import numpy as np
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt

vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(farmers)), labels=farmers)
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```

# Scatter plots

**Purpose**: Displaying relationships between variables.

**matplotlib function**: [`scatter(x, y)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter)

* `x`, `y`: The values for the two variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("data/iris_csv.csv") 

fig, ax = plt.subplots()

ax.scatter(iris['petallength'], iris['petalwidth'])
ax.grid(True)
plt.show()
```

* we can represent multiple superimposed plots, according to a grouping (based on attributes)

```python
groups=iris.groupby("class");
labcount=groups.size()
s=labcount.size;
fig, ax = plt.subplots(1, figsize=(10, 6)) #set size of the image
fig.suptitle('Example Of Scatterplot')

for name, group in groups:
    ax.plot(group.sepallength, group.sepalwidth, marker='o', linestyle='', ms=8, label=name)
ax.legend()

#loops. automatically set colors according to sequence
```

* We don't need to iterate to separate classes

```python
fig, ax = plt.subplots()
colors = {'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'blue'}
ax.scatter(iris['petallength'], iris['petalwidth'],c=iris['class'].map(colors))
ax.grid(True)
plt.show()
```
